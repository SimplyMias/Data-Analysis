#The NumPy ndarray (A multidimensional Array object)

arrays enables us to perform mathematical operations on whole block of data

```python
import numpy as np
data = np.array([[1.5,-0.1,3],[0,-3,6.5]])
```

```python
data
```

mathematical operation with data

```python
data * 10
```

```python
data + data
```

In the first example, all of the elements have been multiplied by 10. In the second, the
corresponding values in each “cell” in the array have been added to each other.

An ndarray is a generic multidimensional container for homogeneous data; that is, all of the elements must be the same type

Every array has a shape, a tuple indicating the size of each dimension and a

dtype, an object describing the data type of the array

```python
data.shape #tuple indicating size of each dimension
```

```python
data.dtype #an object describing the data type of array
```

**Creating ndarrays**

array function can be used to create an array

array function can accept any sequence-like oject (including other arrays)

```python
data1 = [6, 7.5, 8, 0, 1]
```

```python
array1 = np.array(data1)
```

```python
array1
```

*Nested sequences* like a list of equal length lists will be converted into multidimensional array

```python
data2 = [[1,2,3,4],[5,6,7,8]]
```

```python
arr2 = np.array(data2)
```

```python
arr2
```

Since data2 was a list of lists, the NumPy array arr2 has two dimensions, with
shape inferred from the data. We can confirm this by inspecting the ndim and shape
attributes:

```python
arr2.ndim
```

```python
arr2.shape
```

numpy.array tries to infer a good data type for the array that it creates. The data
type is stored in a special dtype metadata object; for example, in the previous two
examples we have:

```python
array1.dtype
```

```python
arr2.dtype
```

**Data Types for ndarrays**

A data type or dtype is special object containing the information (or metadata, data about data)

```python
arr1 = np.array([1,2,3], dtype = np.float64)
arr2 = np.array([1,2,3], dtype = np.int64)
```

```python
arr1.dtype
```

```python
arr2.dtype
```

we can explicitly convert or cast an array from one data type to another using ndarray's astype method

```python
arr = np.array([1,2,3,4,5])
```

```python
arr.dtype
```

```python
float_arr = arr.astype(np.float64)
```

```python
float_arr
```

```python
float_arr.dtype
```

In this example integers are casted or converted to floats but if we convert floats to integers the decimal part will be truncated

```python
arr = np.array([3.7, -1.2, -2.6, 0.5, 12.9, 10.1])
```

```python
arr
```

```python
arr.astype(np.int32)
```

astype function can also be used if we have strings representing numbers we can explicitly convert it to the numeric form

```python
numeric_string = np.array(["1.25", "-9.6", "42"], dtype = str )
```

```python
numeric_string.astype(float)
```

If casting were to fail for some reason (like a string that cannot be converted to
float64), a ValueError will be raised. Before, I was a bit lazy and wrote float
instead of np.float64; NumPy aliases the Python types to its own equivalent data
types

You can also use another array’s dtype attribute:


```python
int_array = np.arange(10 )
```

```python
calibers = np.array([.22, .270, .357, .380, .44, .50], dtype=np.float64)
```

```python
int_array.astype(calibers.dtype)
```
